```python
# Load a csv file on pandas
#Load csv file dataset on pandas 

df = pd.read_csv("data/jiji_housing_raw.csv")

df
```

```python
df.info()
```

```python
#Missing Values

df.isnull().sum()
```

```python
# I dropped all the null values in the dataset since its just 4 data out of 2000 dataset. 

df.dropna(inplace=True)

# Rechecking the null values in the dataset

df.isnull().sum()
```

```python
df.shape
```

```python
df.columns
```

```python
#Data cleaning 
#Question: Standardize region, region_name, region_parent_name.

df.rename(columns={
    "title" : "Title" , 
    "property_size" : "Property Size" ,
    "bedrooms" : "Bedrooms" ,
    "bathrooms" : "Bathrooms" ,
    "furnishing" : "Furnishing" ,
    "region" : "Region" ,
    "region_name" : "Region Name" ,
    "region_parent_name" : "Region Parent Name",
    "is_boost" : "Is Boost",
    "price" : "Price"
} , inplace=True)

df.head(3)
```

```python
#check for unique values in each colunms

print(df["Title"].unique())
print(df["Property Size"].unique())
print(df["Bedrooms"].unique())
print(df["Bathrooms"].unique())
print(df["Furnishing"].unique())
print(df["Region"].unique())
print(df["Region Name"].unique())
print(df["Region Parent Name"].unique())
print(df["Is Boost"].unique())
print(df["Price"].unique())
```

```python
# Price cleaning 
# Question: Convert the price from strings like “₦ 85,500,000” to float (85500000.00)

df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace("₦", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
df["Price"]
```

```python
#Are there outlier prices that should be removed?
#exclude an outliers in dataframe

Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
Q1
Q3
IQR = Q3 -Q1
IQR

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ( df['Price'] < lower_bound) | ( df['Price'] > upper_bound)

print(outliers.sum())
```

```python
# Create a new df after removing price column outliers 

new_df = df[
    (df['Price'] > lower_bound) & ( df['Price'] < upper_bound)
] 

new_df
```

```python
# Question: Convert furnishing type to consistent categories (e.g., Furnished / Semi-furnished / Unfurnished / Unknown)

new_df["Furnishing"] = (
    new_df["Furnishing"]
    .str.lower()
    .replace({
        "semi furnished": "Semi-Furnished",
        "furnished": "Furnished",
        "unfurnished": "Unfurnished"
    })
)

new_df
```

```python
new_df["Region Parent Name"].unique()
```

```python
new_df['Region Parent Name'] = new_df['Region Parent Name'].replace(['Ajah', 'Ibeju', 'Ikorodu', 'Magodo', 'Lekki', 'Delta State', 'Ikeja', 'Gbagada',
                                                                     'Ikotun/Igando', 'Ejigbo','Surulere', 'Isolo', 'Ojodu', 'Egbeda', 'Ogudu',
                                                                     'Maryland', 'Kosofe', 'Alimosho', 'Ikoyi', 'Egbe/Idimu', 'Agege', 'Ondo State',
                                                                     'Victoria Island', 'Ipaja', 'Amuwo-Odofin', 'Ogba', 'Lagos Island (Eko)','Oshodi',
                                                                     'Mushin', 'Shomolu', 'Ifako-Ijaiye','Ilupeju', 'Epe'], 'Lagos State')

new_df['Region Parent Name'].unique()
```

```python
new_df['Region Parent Name'] = new_df['Region Parent Name'].replace(['Lugbe District', 'Katampe', 'Apo District','Wuse',
                                                                     'Gwagwa', 'Gwarinpa', 'Garki 1', 'Jiwa', 'Bwari',  ], 'Abuja (FCT)')

new_df['Region Parent Name'].unique()
```

```python
new_df['Region Parent Name'] = new_df['Region Parent Name'].replace(['Ibadan'], 'Oyo State')


new_df['Region Parent Name'].unique()
```

```python
new_df['Region Parent Name'] = new_df['Region Parent Name'].replace(['Sagamu'], 'Ogun State')

new_df['Region Parent Name'].unique()
```

```python
# Question: Remove duplicates;Look for duplicate titles or identical rows.
new_df.drop_duplicates(inplace=True)
new_df
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
# Question: Save cleaned dataset as:  jiji_housing_cleaned.csv

new_df.to_csv("data/jiji_housing_cleaned.csv", index=False)
```

```python
#Task 3: Exploratory Data Analysis (EDA)
# 1. Price Analysis
#Question: What is the average house price in Nigeria?

avg_house_price = round(new_df["Price"].mean(), 2)


print(f"Average House Price:", avg_house_price)
```

```python
#Question: Which state has the highest and lowest mean property price?

state_property_price = (
    new_df.groupby("Region Parent Name")["Price"]
    .mean()
    .sort_values(ascending=False)
)
print("State With Highest Mean Property Price:")
print(state_property_price.head())

print("\nState With Lowest Mean Property Price:")
print(state_property_price.tail())
```

```python
# 2. Property Size & Features
# Question: What is the distribution of property sizes?

new_df["Property Size"].value_counts().head(10)
```

```python
## Question: Do houses with more bedrooms/bathrooms cost significantly more?
region_counts = new_df["Region Name"].value_counts()

print(region_counts.head(10))
```

```python
# 4. Furnishing Analysis
# Question: Are furnished apartments more expensive on average?

furnishing_aprt = (
    new_df.groupby("Furnishing")["Price"]
    .mean()
    .sort_values(ascending=False)
)

print(furnishing_aprt)
```

```python
# 5. Listing Type
# Do boosted/enterprise listings have higher prices?
boost_price = (
    new_df.groupby("Is Boost")["Price"]
    .mean().sort_values(ascending=False)
)

print(boost_price)
```

```python
#Correlation Analysis
corr = new_df[
    [
        "Price",
        "Property Size",
        "Bedrooms",
        "Bathrooms"
    ]
].corr()

corr
```
